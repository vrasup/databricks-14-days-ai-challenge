```python
# Load Data
path = "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv"
df = spark.read.csv(path, header = True, inferSchema = True)
df.show(5)
```

```python
#Basic Operations
#Select columns
df.select("event_type", "product_id", "price").show(10)
```

```python
#Filter by price>100
df.filter("price >100").show(5)
```

```python
#Group by event type
df.groupBy("event_type").count().show()
```

```python
#Top 5 brands
top_brands = df.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
top_brands.show(5)
```
