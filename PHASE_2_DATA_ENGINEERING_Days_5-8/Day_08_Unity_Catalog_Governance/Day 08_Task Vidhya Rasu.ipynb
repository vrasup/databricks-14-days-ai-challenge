#### Step 2: Create Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

#### Step 3: Register Bronze table

```python
%sql
DROP TABLE IF EXISTS bronze.events;

CREATE TABLE bronze.events AS
SELECT * FROM workspace.default.bronze_df_nov;

SELECT COUNT(*) FROM bronze.events;
```

#### Step 4: Register SILVER table

```python
%sql
DROP TABLE IF EXISTS silver.events_clean;

CREATE TABLE silver.events_clean AS
SELECT * FROM workspace.default.silver_df_nov_realworld;

SELECT COUNT(*) FROM silver.events_clean;
```

#### Step 5:Register GOLD tables

```python
%sql
DROP TABLE IF EXISTS gold.product_metrics;
DROP TABLE IF EXISTS gold.category_metrics;
DROP TABLE IF EXISTS gold.daily_metrics;

CREATE TABLE gold.product_metrics AS
SELECT * FROM workspace.default.gold_df_product_nov;

CREATE TABLE gold.category_metrics AS
SELECT * FROM workspace.default.gold_df_category_nov;

CREATE TABLE gold.daily_metrics AS
SELECT * FROM workspace.default.gold_df_daily_nov;

SHOW TABLES IN gold;
```

#### Step 6: Create GOVERNED View

```python
%sql
CREATE OR REPLACE VIEW gold.top_products AS
SELECT
  product_id,
  brand,
  revenue,
  conversion_rate
FROM gold.product_metrics
WHERE revenue > 0
ORDER BY revenue DESC
LIMIT 100;

SELECT * FROM gold.top_products;
```

#### Step 7: Lineage Proof

```python
%sql
DESCRIBE EXTENDED gold.top_products;
```
